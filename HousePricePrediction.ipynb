```python
from sklearn.linear_model import LinearRegression
import pandas as pd

# Load the dataset
data = pd.read_csv('Housing.csv')

# Handle missing values separately for numeric and categorical columns
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns
categorical_columns = data.select_dtypes(include=['object']).columns

# Fill missing values for numeric columns with the median
data[numeric_columns] = data[numeric_columns].fillna(data[numeric_columns].median())

# Fill missing values for categorical columns with the mode (most frequent value)
for col in categorical_columns:
    data[col] = data[col].fillna(data[col].mode()[0])

# Feature Selection and Preprocessing
X = data[['area', 'bedrooms', 'mainroad']]  # Example of features (you can add others if needed)
y = data['price']  # Target variable

# Encode categorical columns (if any)
X['mainroad'] = X['mainroad'].apply(lambda x: 1 if x == 'yes' else 0)

# Initialize and train the model
model = LinearRegression()
model.fit(X, y)

# Now the model is trained and can be used for predictions
def predict_house_price_simple(model, data):
    # Find the min and max values for area and bedrooms
    min_area = data['area'].min()
    max_area = data['area'].max()
    min_bedrooms = data['bedrooms'].min()
    max_bedrooms = data['bedrooms'].max()

    # Ask for user input
    area = float(input(f"Enter the area (min={min_area}, max={max_area}): "))
    bedrooms = int(input(f"Enter the number of bedrooms (min={min_bedrooms}, max={max_bedrooms}): "))
    mainroad = input("Is the house near the main road? (yes/no): ").strip().lower()

    # Ensure valid mainroad input (yes/no)
    if mainroad not in ['yes', 'no']:
        print("Invalid input for mainroad. Please enter 'yes' or 'no'.")
        return

    # Map mainroad to numeric value (1 for 'yes', 0 for 'no')
    mainroad_value = 1 if mainroad == 'yes' else 0

    # Prepare the input data (normalize area and bedrooms)
    normalized_area = (area - min_area) / (max_area - min_area)
    normalized_bedrooms = (bedrooms - min_bedrooms) / (max_bedrooms - min_bedrooms)

    # Create a DataFrame for prediction
    user_data = pd.DataFrame({
        'area': [normalized_area],
        'bedrooms': [normalized_bedrooms],
        'mainroad': [mainroad_value]
    })

    # Predict the house price
    predicted_price = model.predict(user_data)

    # Display the predicted price
    print(f"The predicted house price is: {predicted_price[0]}")

# After training the model, call the function
predict_house_price_simple(model, data)
```
